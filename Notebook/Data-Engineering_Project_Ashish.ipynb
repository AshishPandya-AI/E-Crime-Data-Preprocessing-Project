### **2.Missing at Random (MAR)**

```python
from scipy.stats import chi2_contingency

# Add a column to indicate missingness in 'Weapon'
missing_col = 'Weapon'
df['Missing_Weapon'] = df[missing_col].isna().astype(int)

# List of categorical columns to test
categorical_columns = ['Inside/Outside', 'District', 'Premise']

# Store Chi-Square results
chi_square_results = {'Attribute': [], 'Chi2_Statistic': [], 'P_Value': []}

for col in categorical_columns:
    # Create a contingency table
    contingency_table = pd.crosstab(df['Missing_Weapon'], df[col])

    # Perform Chi-Square Test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Print results
    print(f"Chi-Square Test for {col} and Missing_Weapon:")
    print(f"Chi2 = {chi2}, p-value = {p}")

    # Add results to the dictionary
    chi_square_results['Attribute'].append(col)
    chi_square_results['Chi2_Statistic'].append(chi2)
    chi_square_results['P_Value'].append(p)
```

```python
# Convert results to a DataFrame
chi_square_df = pd.DataFrame(chi_square_results)

# Plot heatmap for Chi2 Statistic
plt.figure(figsize=(8, 4))
sns.heatmap(
    chi_square_df[['Chi2_Statistic']].set_index(chi_square_df['Attribute']),
    annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={'label': 'Chi2 Statistic'}
)
plt.title('Chi-Square Statistics Heatmap')
plt.ylabel('Attribute')
plt.show()
```

```python
# Plot heatmap for P-Value
plt.figure(figsize=(8, 4))
sns.heatmap(
    chi_square_df[['P_Value']].set_index(chi_square_df['Attribute']),
    annot=True, fmt=".2e", cmap="YlGnBu", cbar_kws={'label': 'P-Value'}
)
plt.title('Chi-Square P-Values Heatmap')
plt.ylabel('Attribute')
plt.show()
```

```python
# Define a condition to detect outliers
is_outlier = (df['Number of Witnesses'] < 1) | (df['Number of Witnesses'] > 15)

# Create labels for box plot
labels = ['Normal Range', 'Outliers']

# Separate the data into two groups: normal range and outliers
box_data = pd.Series('', index=is_outlier.unique())
for condition in is_outlier.unique():
    mask = is_outlier == condition
    box_data[condition] = df[mask]['Number of Witnesses']

# Plot the box plot
plt.boxplot(box_data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.yticks([1, 2], labels)
plt.title('Box Plot for Number of Witnesses')
plt.xlabel('Number of Witnesses')
plt.show()
```

```python
# Define the acceptable range
lower_limit = 1
upper_limit = 15

# Handle outliers: Replace values outside the range with the boundary values
df['Number of Witnesses'] = df['Number of Witnesses'].clip(lower=lower_limit, upper=upper_limit)

# Define a condition to check for original outliers before capping
is_outlier = (df['Number of Witnesses'] < lower_limit) | (df['Number of Witnesses'] > upper_limit)

# Create labels for box plot
labels = ['Normal Range (After Fix)', 'Original Outliers']

# Separate the data into two groups: normal range and original outliers
box_data = pd.Series('', index=is_outlier.unique())
for condition in is_outlier.unique():
    mask = is_outlier == condition
    box_data[condition] = df[mask]['Number of Witnesses']

# Plot the box plot
plt.boxplot(box_data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.yticks([1, 2], labels)
plt.title('Box Plot for Number of Witnesses (After Handling Outliers)')
plt.xlabel('Number of Witnesses')
plt.show()
```

**1. Why I Fixed Outliers for the Number of Witnesses**

* **Consistency:** I fixed outliers to ensure the data is consistent and reflects realistic scenarios. This is important for any further analysis or machine learning model.
* **Accurate Analysis:** By addressing outliers, we can ensure that summary statistics and models are based on valid data, leading to more reliable insights.
* **Improved Model Performance:** Cleaning outliers improves model performance by making the feature more meaningful and preventing skewed learning from rare, extreme data points.

**2. How I Decided Which Outliers to Fix**

* I identified extreme outliers based on statistical methods, like visualizing the distribution (e.g., box plots or histograms) or using methods like Z-scores or IQR (Interquartile Range).
* After identifying the outliers, I applied a reasonable strategy (e.g., capping, removal, or imputation) based on the nature of the data and domain knowledge about what constitutes a realistic number of witnesses for a crime.

## **Data profile:-**

```python
# Initialize the data definition and profile list
data_profile = []

# Define the data definition and profiling function
def create_data_profile(df):
    for column in df.columns:
        data_type = df[column].dtype  # Data type of the column
        unique_values = df[column].nunique()  # Count of unique values
        missing_values = df[column].isna().sum()  # Count of missing values
        missing_percentage = (missing_values / len(df)) * 100  # Percentage of missing values

        if pd.api.types.is_numeric_dtype(df[column]):
            col_min = df[column].min()  # Minimum value
            col_max = df[column].max()  # Maximum value
            col_mean = df[column].mean()  # Mean value
            col_std = df[column].std()  # Standard deviation
        else:
            col_min = col_max = col_mean = col_std = None

        # Append the column's profile to the list
        data_profile.append({
            "Attribute": column,
            "Data Type": data_type,
            "Unique Values": unique_values,
            "Missing Values": missing_values,
            "Missing Percentage": f"{missing_percentage:.2f}%",
            "Min": col_min,
            "Max": col_max,
            "Mean": col_mean,
            "Std Dev": col_std
        })

# Call the function to populate the data profile
create_data_profile(df)

# Convert the profile list to a DataFrame for better readability
profile_df = pd.DataFrame(data_profile)

# Display the data profile
profile_df
```
